```python
def map_rating(score):
    if pd.isnull(score):  
        return ''  
    if score >= 3.5 and score <= 5:
        return 'high'
    elif score >= 2.5 and score < 3.5:
        return 'neutral'
    else:
        return 'low'

for column in df.select_dtypes(include='object').columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# 应用函数到每一列，假设所有的列都需要转换评级
df = df.applymap(map_rating)

print(df)
```

```python
f = open('rating_test.pkl','wb')
pickle.dump(df, f)
f.close()
```
